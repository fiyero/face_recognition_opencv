## how to train a classifier for face recognization

```python
import cv2
import os
from PIL import Image
import numpy as np
import pickle

image_dir="images"
face_cascade= cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
recognizer=cv2.face.LBPHFaceRecognizer_create()

x_train=[]
y_train=[]
current_id=0

label_ids={}

for root,dirs,files in os.walk(image_dir):
    for file in files:
        if file.endswith('png') or file.endswith('jpg'):
            path=os.path.join(root,file)  
            label=os.path.basename(os.path.dirname(path)).replace(" ","-").lower()

            
            #create label id for labelling as we cannot feed text as label into model
            if label not in label_ids:
                label_ids[label]=current_id
                current_id+=1


            pil_image=Image.open(path).convert("L") #convert the color image into gray
            
            size=(550,550)
            final_image = pil_image.resize(size, Image.ANTIALIAS)
            image_array=np.array(final_image,'uint8')

            faces= face_cascade.detectMultiScale(image_array,scaleFactor=1.5,minNeighbors=3 )
            for (x,y,w,h) in faces:
                roi_gray=image_array[y:y+h,x:x+w]
                x_train.append(roi_gray)
                #so the face information of the img in np array is appended into our training data
                y_train.append(label_ids[label])


with open('labels.pickle', 'wb') as f:
    
    pickle.dump(label_ids, f)
    
recognizer.train(x_train,np.array(y_train))
recognizer.write('trainner.yml')
```
